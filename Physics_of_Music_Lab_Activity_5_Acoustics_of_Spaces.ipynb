#<font color='red'>Read in and Visualize the Data

<font color='red'>When we're finished, the code cell below will read in the data from this spreadsheet and then produce a graph of the sound levels that we measured at each location in the hall.

```python
!pip install openpyxl
!pip install xlrd
import pandas as pd
my_sheet = 'data'
url = 'https://docs.google.com/spreadsheets/d/1yHH_XOQ3IhNl2zMnTrcdvklW4cx3FyhFyO4lc-Q8FXA/export'
dataframe = pd.read_excel(url, sheet_name=my_sheet,engine='openpyxl')

import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(dataframe['x'], dataframe['y'], dataframe['sound level (decibels)'], marker='o')

ax.view_init(30, 30) # You can change these numbers to rotate the view.
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(dataframe['x'], dataframe['y'], dataframe['sound level (decibels)'], marker='o')
ax.view_init(0, -90) # You can change these numbers to rotate the view.
plt.show()
# The first angle in view_init() moves the view up or down.
# The second angle in view_init() rotates the view left and right.
# There's a very high point in the back
# just so we know which side is which!
```

```python
import numpy as np
print('average decibels = ',np.mean(dataframe['sound level (decibels)'][0:len(dataframe)-1]))
print('decibels standard deviation= ',np.std(dataframe['sound level (decibels)'][0:len(dataframe)-1]))
print('decibels percent standard deviation= ',np.std(dataframe['sound level (decibels)'][0:len(dataframe)-1])/np.mean(dataframe['sound level (decibels)'][0:len(dataframe)-1])*100,'%')
```
