# Make notes

The code cell below creates tones of given frequencies and then overlays them onto each other.

```
import sys
sys.path.insert(0, 'ThinkDSP/code/')
import thinkdsp
import matplotlib.pyplot as pyplot
import IPython

# 1. Create a wave for each frequency.
s1 = thinkdsp.SinSignal(freq=500, amp=0.5)
s2 = thinkdsp.SinSignal(freq=1000, amp=0.5)
# Copy-paste to create more signals!

# 2. Create the wave.
mix1 = s1+s2
wave1 = mix1.make_wave(duration=2, start=0, framerate=44100)

# 3. Play wave
wave1.play()
w1 = IPython.display.Audio('sound.wav')
display(w1)
```
