#<font color='red'>Creating a Model

<font color='red'>We determined in class that this data looks like it could fit a **grid pattern**, which is what we would expect for a standing wave in this space. The code cell below creates a grid with vertical lines at a series of locations called `x_lines` and horizontal lines at a series of locations called `y_lines`.

#<font color='red'>Distances Between Lines

<font color='red'>The spacing on your grid should be related to the wavelength of the standing wave. In order to determine this wavelength, we need to examine the distances between the lines, or the **grid spacing**. The code cell below calculates the distances between neighboring vertical lines and stores those distances in a list called `distances`.

<font color='red'>Explain how this code uses a [for loop](https://www.w3schools.com/python/python_for_loops.asp) to carry out this process.

<font color='red'>Create a copy of the for loop in this cell and modify the new copy to calculate and store the distances between neighboring horizontal lines.

```python
# Calculate distances between nearest lines.
distances = []
for i in range(1,len(x_lines)):
  distances.append(x_lines[i]-x_lines[i-1])


print(distances)
```

```python
bins = [55,60,65,70,75,80,85,90]
plt.hist(distances,bins)
```

#<font color='red'>Calculate Average and Standard Deviation

<font color='red'>Viewing a histogram is nice, but we need to easily summarize our data. One way you can summarize a large set of numerical data is to report a **mean** and a **standard deviation**. A mean is a measure of the central value in the histogram: It returns an average value that has equal "weighting" above and below). A standard deviation is a measure of how broad the distribution of values is on the histrogram. If the standard deviation is **small compared to the mean**, then we have pretty good confidence in our measurements.

<font color='red'>The code cell below calculates the mean and standard deviation of our length measurements. Run the code cell and discuss whether you would consider this standard deviation small compared to the mean.

```python
import numpy as np
print(np.average(distances))
print(np.std(distances))
print(np.std(distances)/np.average(distances)*100)
```
